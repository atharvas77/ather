```python
import matplotlib.pyplot as plt
```

```python
def output(x):
    return((x+3)**2)

def derivate(x):
    return (2*x+6)

def main():
    x=2;
    steps=200
    alpha=0.1

    x_values=[]
    y_values=[]

    while(steps>0):
        
        previous_value=output(x)
        x_values.append(x)
        y_values.append(previous_value)

        print("x={:.10f}, y={:.10f}".format(x,previous_value))

        x=x-alpha*derivate(x)

        if(output(x)>previous_value):
            x=x+alpha*derivate(x)
            break

        steps-=1

    print()
    print("Local minima occurs at : x:{:.10f} , y:{:.10f}".format(x,previous_value))

    plt.scatter(x_values,y_values,color="magenta")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Gradient Descent")
                
main() 
```
